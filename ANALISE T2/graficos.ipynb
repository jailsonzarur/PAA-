```python
import matplotlib as plt
```

```python


# Dados de exemplo
y1 = [0.00467, 0.028136, 0.060686, 0.381766]
y2 = [0.005244, 0.015042, 0.0346102, 0.161852]
x = [100, 500, 1000, 5000]

plt.figure(figsize=(10, 6))
# Criando o gráfico de linhas com marcadores
plt.plot(x, y1, marker='o', label='Heap')  # 'o' indica o estilo do marcador (círculo)
plt.plot(x, y2, marker='o', label='Quick')  # 'o' indica o estilo do marcador (círculo)

# Adicionando título e rótulos
plt.title('Tempo Médio de Execução x Tamanho do Vetor')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio de Execução(Em milisegundos)')
plt.grid(visible=True)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Mostrando o gráfico
plt.show()
```

```python
# Dados de exemplo
y1 = [0.00467, 0.028136, 0.060686, 0.381766, 3.56121, 6.17958, 9.46487, 18.1501, 20.2902]
y2 = [0.005244, 0.015042, 0.0346102, 0.161852, 0.8956, 2.73, 3.32, 4.95, 6.46]
x = [100, 500, 1000, 5000, 30000, 80000, 100000, 150000, 200000]

plt.figure(figsize=(10, 6))
# Criando o gráfico de linhas com marcadores
plt.plot(x, y1, linestyle='-', label='Heap')  # 'o' indica o estilo do marcador (círculo)
plt.plot(x, y2, linestyle='-', label='Quick')  # 'o' indica o estilo do marcador (círculo)

# Adicionando título e rótulos
plt.title('Tempo Médio de Execução x Tamanho do Vetor')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio de Execução(Em milisegundos)')
plt.grid(visible=True)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Mostrando o gráfico
plt.show()
```
